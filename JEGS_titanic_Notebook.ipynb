```python
import pandas as pd
import numpy as np
import re
```

```python
#Impotacion de  datos
df_titanic = pd.read_csv('https://raw.githubusercontent.com/omarmendoza564/datos/main/datos/titanic.csv')
df_titanic.info()
```

```python
df_titanic
```

# 1. Si la edad es null (na) llenar el campo con la media de la edad (usar el método fillna())

```python
# Comprabamos que hay elementos nulos
df_nan = df_titanic[pd.isnull(df_titanic['age'])]['age'].isna()
df_nan
```

```python
# Usamos fillna para remplazar los elementos con valor nulo
df_titanic['age'].fillna((df_titanic['age'].mean()), inplace=True)
```

```python
# Comprobamos que ya no hay valores nulos en edad
print(df_titanic['age'].isnull().sum())
print(df_nan.index)
df_titanic[df_titanic.index.isin(df_nan.index)]
```

# 2. Si embarqued es na llenar con S

```python
# Comprabamos que hay elementos nulos
df_nan = df_titanic[pd.isnull(df_titanic['embarked'])]['embarked'].isna()
df_nan
```

```python
df_titanic['embarked'].fillna('S', inplace=True)
```

```python
# Comprobamos que los elementos se cambaiaran a (S) 
print(df_titanic['embarked'].isnull().sum())
print(df_nan.index)
df_titanic[df_titanic.index.isin(df_nan.index)]
```

# 3. Cambiar los valores alfanumericos de sex a numero usando { 'female': 1, 'male': 0 }

```python
df_titanic.info()
```

```python
#Usamos el metodo replace para cambiar remplazar los valores
df_titanic['sex'] = df_titanic['sex'].replace(['female','male'],[1,0])
```

```python
df_titanic['sex']
```

# 4. Cambiar los valores alfanumericos de embarked a numero usando { 'S': 0, 'C': 1, 'Q': 2 }

```python
df_titanic.info()
```

```python
#Usamos el metodo replace para cambiar remplazar los valores
df_titanic['embarked'] = df_titanic['embarked'].replace(['S','C','Q'],[0,1,2])
```

```python
df_titanic['embarked']
```

# 5. Crear una nueva variable llamada child con valor 1 para los pasajeros con una edad menor a 16 años y 0 si el pasajero tiene 16 años o mas

```python
#Creamos una nueva columna y usando where determinamos nuestra condicion

df_titanic['child'] = np.where(df_titanic['age'] >=16, 0,1 )
```

```python
df_titanic['child']
```

# 6. Eliminar las columnas ticket, cabin y sibsp

```python
# Eliminamos las columnas ticket, cabin y sibsp del dataframe
df_titanic = df_titanic.drop(['ticket','cabin','sibsp'],axis=1)
```
