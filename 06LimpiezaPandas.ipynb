#Limpieza de datos con Pandas

```python
import pandas as pd
import numpy as np
import datetime as dt
import re
```

```python
#Leer origen de datos
df = pd.read_csv('https://raw.githubusercontent.com/omarmendoza564/datos/main/datos/aerolineas.csv')
df.info()
```

```python
df.head()
```

```python
df.tail()
```

```python
#Visualizar el numero de datos faltantes
df.isnull().sum()
```

```python
df_nan =df[pd.isnull(df['espera_min'])]['espera_min'].isna()
df_nan
```

```python
#Ver los datos descriptivos de espera en minutos
df['espera_min'].describe()
```

```python
df.iloc[2227]
```

```python
#Cambiar los valores NaN por el promedio de espera_min
df['espera_min'].fillna((df['espera_min'].mean()), inplace=True)
```

```python
df.isnull().sum()
```

```python
#Visualizar los cambios de valores NaN
print(df_nan.index)
df[df.index.isin(df_nan.index)]
```

```python
#Eliminar columnas
df.drop(['nrow'],axis=1)
```

```python
#Crear una columna
df['espera_hr'] = df['espera_min'] / 60
df[['id','aerolinea','espera_min','espera_hr']]
```

```python
df.info()
```

```python
#Convertir columnas tipo cadenas(object) a tipo date
df['fecha_salida'] = pd.to_datetime(df['fecha_salida']).dt.date
df
```

```python
print(df['fecha_salida'].describe())
print("Fecha minima ", df['fecha_salida'].min())
print("Fecha maxima ", df['fecha_salida'].max())
```

```python
#Filtar renglones por fecha
df.loc[df['fecha_salida'] == dt.date(year=2018, month=12, day=31)]
```

```python
#Localizar renglones duplicados
df_dup = df.duplicated()
df_dup
```

```python
#Localizamos los renglones duplicados
df[df_dup]
```

```python
#Definir columnas a evaluar duplicado
col = ['id', 'dia', 'aerolinea']
df_dup = df.duplicated(subset=col, keep = False)
df[df_dup].sort_values(by= 'id')
```

```python
#Eliminar duplicados
df.drop_duplicates(inplace = True)
df.count()
```

```python
#Cargar el DF de categorias
df_cat = pd.read_csv('https://raw.githubusercontent.com/omarmendoza564/datos/main/datos/categorias.csv')
df_cat
```

```python
#Visualizar los valores unicos de cada columna categorica (limpieza, seguridad, satisfaccion )
# unique()
print('Limpieza', df['limpieza'].unique())
print('Seguridas', df['seguridad'].unique())
print('Satisfaccion', df['satisfaccion'].unique())
```

```python
# Localizar los valores invalidos en la columna limpieza
df_limpieza = set(df['limpieza']).difference(df_cat['limpieza'])
df_limpieza
```

```python
#Localizar renglones que no tienen un valor valido para limpienza
df_lim_r = df['limpieza'].isin(df_limpieza)
df[df_lim_r]
```

```python
#Contar renglones que tienen valores invalidos en limpieza
df[df_lim_r].count()
```

```python
#Contar renglones que tienen valores validos en limpieza
df[~df_lim_r].count()
```

```python
#Reemplazar valores en una columna
df['limpieza'] = df['limpieza'].replace({'Inaceptable':"Sucio", 'Medio':'Promedio'})
set(df['limpieza']).difference(df_cat['limpieza'])
#Validar columnas region, dest_size
print('Region', df['region'].unique())
print('Dest Size', df['dest_size'].unique())
```

```python
#Transformar datos

#Pasar a minusculas
df['region'] = df['region'].str.lower()

#Reemplazar valores
df['region'] = df['region'].replace({'eur':'europa'})

#Quitar espacios en blanco
df['dest_size'] = df['dest_size'].str.strip()

#Usar expresiones regulares para reemplazo
df['dest_size'] = df['dest_size'].str.replace("grande", "Large", regex=True, flags=re.IGNORECASE)
```

```python
#Guardar DF
df.to_csv('datos/df/aerolinea_limpio.cvs',index=False)
```
